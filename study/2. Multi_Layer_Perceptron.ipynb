```python
import os
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

# 2. 다층 신경망 구현 #

  + 입력 layer와 출력 layer 사이에 여러 개의 layer가 존재하는 구조
  + High Level API를 이용하여 단순하게 구현 가능

neuralnet_mlp_1.png

a. 함수를 정의한다.

```python
def function(x):
  return 4*x**3 + 1.5*x**2 + 0.7*x + 1.6
```

b. train 데이터를 생성하고 시각화한다.

```python
import random

# train 데이터 생성
train_x = []
train_y = []

# -10 ~ 10 사이의 난수 x
for i in range(1000):
  random_x = -10 + 20 * random.random()
  train_x.append([random_x])
  train_y.append([function(random_x)])

train_x
```

```python
# 그래프 표시
plt.plot(train_x, train_y, 'o')
plt.show()
```

c. 모델을 생성, 컴파일 및 학습을 진행한다.

```python
# 데이터 준비
train_x = np.array(train_x)
train_y = np.array(train_y)

# 모델 생성
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(shape=(1,)))
model.add(tf.keras.layers.Dense(64, activation = "relu"))
model.add(tf.keras.layers.Dense(64, activation = "relu"))
model.add(tf.keras.layers.Dense(1))

# 모델 컴파일
model.compile(optimizer='adam', loss='mse')

# 모델 학습
model.fit(train_x, train_y, epochs=50, batch_size=32, verbose=1)
```

d. test데이터를 생성하고 예측한다.

```python
# test 데이터 생성
test_x = []
test_y = []

# -10 ~ 10 사이의 난수 x
for i in range(100):
  random_x = -10 + 20 * random.random()
  test_x.append([random_x])
  test_y.append([function(random_x)])

# 배열로 생성
test_x = np.array(test_x)
test_y = np.array(test_y)

test_x
```

```python
# 테스트 데이터로 예측
test_predictions = model.predict(test_x)
```

e. 예측 결과를 실제 결과와 비교하고 시각화한다.
  - 그래프를 통해 정의한 함수에 맞게 모델이 잘 학습된 것을 확인할 수 있다.

```python
# 예측 결과와 실제 결과 비교
for i in range(len(test_x)):
    print(f"Input: {test_x[i]}, Predicted Output: {test_predictions[i]}, Actual Output: {test_y[i]}")
```

```python
# 그래프 표시(산점도)

## train 데이터 그래프
plt.scatter(train_x, train_y, label='Train Data')

## test 데이터 그래프
plt.scatter(test_x, test_y, label='Test Data')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Train & Test')
plt.show()
```
